# Dataframes

```python
import numpy as np
import pandas as pd
```

```python
#creating data frame
df=pd.DataFrame(np.random.randint(0,200,20).reshape(5,4),index=['A','B','C','D','E'],columns=['W','X','Y','Z'])
```

```python
df
```

```python
#conditional selection
df>120
```

```python
#filter the data frame according to condition
df[df['W']>120]
```

```python
#filtering dataframe acc to a condition and the slecing column from that
df[df['W']>120]['W']
```

```python
#filtering dataframe based on multiple conditions
df[(df['W']>120 ) | (df['X']>100)]
```

```python
#reseting index
df.reset_index(inplace=True)
df
```

```python
#setting a custom index
df.set_index('index',inplace=True)
df
```

# null values

```python
#creating a dataframe with null values
new=df[df>29]
```

```python
new
```

```python
#dropping missing values from dataframe
new.dropna()
```

```python
#dropping all columns with null values
new.dropna(axis=1)
```

```python
#filling null values with 15
new.fillna(value=15)
```

```python
#filling value with condition
new.fillna(value=df['Y'].sum())
```
